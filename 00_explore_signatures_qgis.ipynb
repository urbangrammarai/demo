# Explore the Spatial Signatures in QGIS

```python
from IPython.display import Image
```

## Download the data

Head over to the [CDRC page of the data product](https://data.cdrc.ac.uk/dataset/spatial-signatures-great-britain) (you will need to setup a free account if you don't have one) and download the geo-file with the signature polygons. We recommend using the GeoPackage but a shapefile is also available:

```python
Image('img/cdrc_download.png', width=750)
```

## Unpack the file

Extract the `.zip` file using your tool of preference:

```python
Image('img/extract.png', width=500)
```

## Open in QGIS

Add the `.gpkg`/`.shp` file to QGIS as a vector layer. When you open it, accept the CRS proposed (the OSGB 1936, the British national grid):

```python
Image('img/qgis_proj.png', width=500)
```

You are now good to go! The data is loaded as a layer in QGIS, and can combine it with other datasets in the same way you would normally.

```python
Image('img/qgis.png', width=500)
```
